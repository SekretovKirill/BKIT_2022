```python
import requests
import matplotlib.pyplot as plt
```

```python
url = "http://localhost:8000/"
def get_data(num):
    return requests.get(url + str(num)).json()
```

```python
y = get_data(20)
x = [i for i in range(1, 21)]
```

```python
fig = plt.figure(figsize = (20, 10))
plt.bar(x, y)
plt.show()
```

```python
fig = plt.figure(figsize = (20, 10))
plt.plot(x, y)
plt.show()
```
